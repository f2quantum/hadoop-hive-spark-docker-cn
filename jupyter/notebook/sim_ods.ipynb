```python
import csv
import random
import datetime
import random
import uuid
import time
import numpy as np
import matplotlib.pyplot as plt

# 数据仿真器

# 总共的轨迹（任务）数量

JOB_NUM = 1000
# 一条轨迹上有100个点
LINE_POINT_SIZE = 1000

# 一小时3600秒
HOUR_HAS_SECONDS = 3600
# 一共有1000台农机 1
MACHINE_NUMBERS = 1
# 整个任务开始时间 2024-01-01 00:00:00
MISSION_START_TIME = 1704038400
# 假设这个农机每1小时运行一次,一次运行1小时
RUN_GAP = 1
# 持续五年 (每天运行4次)
RUN_COUNT = 24 * 365 * 5

# 打点间隔 5s
MAKE_POINT_INTERNAL = 5
# 每小时打点数量
PER_JOB_POINTS = int(3600 / MAKE_POINT_INTERNAL)

job_id = 0

jobs_file = open(f'./data/jobs.csv', 'w', newline="")
jobs_writer = csv.writer(jobs_file)

timeseries_file = open(f'./data/timeseries.csv', 'w', newline="")
timeseries_writer = csv.writer(timeseries_file)

point_cnt = 0
for m in range(MACHINE_NUMBERS):
    current = MISSION_START_TIME
    for cnt in range(RUN_COUNT):
        job_start_timestamp = current
        job_end_timestamp = current + HOUR_HAS_SECONDS
        jobs_writer.writerow([job_id, job_start_timestamp, job_end_timestamp])
        job_id = job_id + 1

        # 45 度 每次走100米
        x = np.linspace(cnt, cnt + 100, PER_JOB_POINTS)  # x轴数据范围

        x = 45 + x * 0.00001141


        slope = 1  # 直线的斜率
        intercept = 1  # 直线的截距
        y = slope * x + intercept  # 直线方程
        # 添加噪音
        noise = np.random.normal(0, 10 * 0.00001141 , len(x))  # 生成服从正态分布的噪音
        y_with_noise = y + noise  # 添加噪音

        current_point = current
        for i in range(PER_JOB_POINTS):
            xi = x[i]
            yi = y[i]
            current_point = current_point + 5

            timeseries_writer.writerow([point_cnt, current_point, xi, yi])
            point_cnt = point_cnt + 1

            if point_cnt % 1000 == 0:
                print(f"Generate {point_cnt} points.")

        current = current + HOUR_HAS_SECONDS
```
